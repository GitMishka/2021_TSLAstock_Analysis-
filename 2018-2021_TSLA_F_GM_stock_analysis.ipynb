```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
%matplotlib inline
import pandas_datareader as web
import datetime
```

```python
#getting data from yahoo api for tesla
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2021,1,1)
tsla = web.DataReader('TSLA','yahoo',start,end)
#tsla_Open=tsla[['Open']]
#tsla_dates=tsla.index.to_frame().reset_index(drop=True)
tsla
```

```python
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2021,1,1)
ford = web.DataReader('F','yahoo',start,end)
ford
```

```python
start = datetime.datetime(2018,1,1)
end = datetime.datetime(2021,1,1)
gm = web.DataReader('GM','yahoo',start,end)
gm
```

```python
#getting data from excel 3 seperate sheets into one sheet, turning date column into index column
T = pd.read_csv('Tesla_Stock.csv',index_col=['Date'],parse_dates=True)
F = pd.read_csv('Ford_Stock.csv',index_col=['Date'],parse_dates=True)
G = pd.read_csv('GM_Stock.csv',index_col=['Date'],parse_dates=True)
ALL = pd.concat([T,F,G])
```

```python
tsla['Open'].plot(label='Tesla',figsize=(16,6),title='Opening Prices')
gm['Open'].plot(label='GM',c='Blue')
ford['Open'].plot(label='Ford',c='Purple')
plt.legend();
# starting 2020 exponential growth in opening price can be observed in TSLA stock
# FORD & GM stocks remain relatively same
```

```python
tsla['Volume'].plot(label='Tesla',figsize=(16,6),title='Volume Traded')
gm['Volume'].plot(label='GM',c='Blue')
ford['Volume'].plot(label='Ford',c='Purple')
plt.legend();
#TSLA and Ford stocks exprience much higher volume trades than GM, which remains steady
#in early 2020 & 2021 TSLA stock has significant rises in volume traded
```

```python
#cumulative daily return
```

```python
tsla['Cumulative Return'] = (1+ tsla['returns']).cumprod()
tsla
#if I invested on 2018-01-02 and kept my investment my cumulative returns would rise over x10 times
```

```python
ford['Cumulative Return'] = (1+ ford['returns']).cumprod()
ford
#if I did same thing for Ford I would actually lose cumulative returns
```

```python
gm['Cumulative Return'] = (1+ gm['returns']).cumprod()
gm
#I would lose cumulative returns if I invested into GM stock, but not as much as Ford stock
```

```python
tsla['Cumulative Return'].plot(label='TSLA',figsize=(16,8),c='Red')
ford['Cumulative Return'].plot(label='F',figsize=(16,8),c='Blue')
gm['Cumulative Return'].plot(label='GM',figsize=(16,8),c='Purple')
plt.legend();
#TSLA promeses exponential cumulative return from 2018 to 2021 time frame
#ford has lowest cumu
```
